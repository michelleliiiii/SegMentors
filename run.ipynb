```python
import json
from pathlib import Path
from typing import Dict, List, Union

import nibabel as nib
import numpy as np


def load_nifti(path: Union[str, Path]) -> np.ndarray:
    """
    Load a .nii.gz file and return the voxel array as integer labels.
    """
    arr = nib.load(str(path)).get_fdata()
    return arr.astype(np.int32)


def dice_for_class(
    pred: np.ndarray,
    gt: np.ndarray,
    class_id: int,
    empty_score: float = 1.0
) -> float:
    """
    Compute Dice score for one class.

    If both prediction and ground truth are empty for this class,
    returns empty_score.
    """
    pred_mask = pred == class_id
    gt_mask = gt == class_id

    pred_sum = pred_mask.sum()
    gt_sum = gt_mask.sum()

    if pred_sum == 0 and gt_sum == 0:
        return empty_score

    intersection = np.logical_and(pred_mask, gt_mask).sum()
    return float((2.0 * intersection) / (pred_sum + gt_sum))


def compare_prediction_folder_to_gt(
    pred_folder: Union[str, Path],
    gt_folder: Union[str, Path],
    output_json: Union[str, Path],
    class_ids: List[int] = None,
    ignore_background: bool = False,
    empty_score: float = 1.0
) -> Dict:
    """
    Compare predicted .nii.gz files with ground-truth .nii.gz files.

    Matching rule:
        pred_folder/case123.nii.gz <-> gt_folder/case123.nii.gz

    Outputs:
        - per-case class-wise Dice
        - mean Dice for each class across all matched cases
        - overall metadata

    Parameters
    ----------
    pred_folder : str or Path
        Folder containing predicted .nii.gz segmentation files.
    gt_folder : str or Path
        Folder containing ground-truth .nii.gz segmentation files.
    output_json : str or Path
        Path to save JSON results.
    class_ids : list[int], optional
        Classes to evaluate. If None, inferred from all matched files.
    ignore_background : bool
        If True and class_ids is inferred automatically, class 0 is removed.
    empty_score : float
        Dice returned when both pred and gt are empty for a class.

    Returns
    -------
    dict
        Results dictionary that is also written to JSON.
    """
    pred_folder = Path(pred_folder)
    gt_folder = Path(gt_folder)
    output_json = Path(output_json)

    pred_files = {f.name: f for f in pred_folder.glob("*.nii.gz")}
    gt_files = {f.name: f for f in gt_folder.glob("*.nii.gz")}

    matched_names = sorted(set(pred_files.keys()) & set(gt_files.keys()))
    pred_only = sorted(set(pred_files.keys()) - set(gt_files.keys()))
    gt_only = sorted(set(gt_files.keys()) - set(pred_files.keys()))

    if len(matched_names) == 0:
        raise ValueError("No matched .nii.gz filenames found between prediction and ground truth folders.")

    # infer class ids if not provided
    if class_ids is None:
        all_classes = set()
        for name in matched_names:
            pred = load_nifti(pred_files[name])
            gt = load_nifti(gt_files[name])

            all_classes.update(np.unique(pred).tolist())
            all_classes.update(np.unique(gt).tolist())

        class_ids = sorted(int(c) for c in all_classes)
        if ignore_background and 0 in class_ids:
            class_ids.remove(0)

    per_case_results = {}
    class_to_scores = {str(c): [] for c in class_ids}

    for name in matched_names:
        pred = load_nifti(pred_files[name])
        gt = load_nifti(gt_files[name])

        if pred.shape != gt.shape:
            raise ValueError(
                f"Shape mismatch for {name}: prediction shape {pred.shape}, ground truth shape {gt.shape}"
            )

        case_result = {}
        for c in class_ids:
            dice = dice_for_class(pred, gt, c, empty_score=empty_score)
            case_result[str(c)] = dice
            class_to_scores[str(c)].append(dice)

        per_case_results[name] = case_result

    mean_dice_per_class = {
        cls: float(np.mean(scores)) if len(scores) > 0 else None
        for cls, scores in class_to_scores.items()
    }

    results = {
        "prediction_folder": str(pred_folder),
        "ground_truth_folder": str(gt_folder),
        "matched_case_count": len(matched_names),
        "matched_cases": matched_names,
        "prediction_only_files": pred_only,
        "ground_truth_only_files": gt_only,
        "evaluated_class_ids": class_ids,
        "empty_class_dice_score": empty_score,
        "per_case_class_dice": per_case_results,
        "mean_dice_per_class": mean_dice_per_class,
    }

    output_json.parent.mkdir(parents=True, exist_ok=True)
    with open(output_json, "w") as f:
        json.dump(results, f, indent=4)

    return results


```

```python

pred_folder = "/content/drive/MyDrive/School/ECE324/code/nnunet/Test_Predictions"
gt_folder = "/content/drive/MyDrive/School/ECE324/code/nnunet/nnUNet_raw/Dataset100_BrainTumor/labelsTs"
output_json = "/content/drive/MyDrive/School/ECE324/code/nnunet/dice_results.json"

# Option 1: specify classes manually
# class_ids = [0, 1, 2, 3]

# Option 2: infer classes automatically
class_ids = None

results = compare_prediction_folder_to_gt(
    pred_folder=pred_folder,
    gt_folder=gt_folder,
    output_json=output_json,
    class_ids=class_ids,
    ignore_background=False,   # set True if you do not want class 0
    empty_score=1.0
)

print("Done.")
print("Mean Dice per class:")
for cls, score in results["mean_dice_per_class"].items():
    print(f"Class {cls}: {score:.4f}")
```
